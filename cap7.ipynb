```python
# Ejercicio 7.2 
nombref = input('por favor diga el nombre del fichero: ')
try:
    x = open(nombref)
    z=0
    t=0
    for i in x:
        if i.startswith('X-DSPAM-Confidence:'):
            z = z + 1
            j = i
            pos_esp = j.find(' ')
            t = t + float(j[pos_esp + 1: ])
    p = t/z 
    print('Número de lineas que empiezan por X-DSPAM-Confidence: son: {}'.format(z))
    print('Valor medio de probabilidad de spam: {}'.format(p))
except: 
    print('verifique el fichero con nombre: {}'.format(nombref))
```
